```python
def vector_len(a, b):
    if a.shape != b.shape:
        return('Нельзя выполнить операцию, вектора имеют разную размерность')
    return math.sqrt(sum([(ai+bj)**2 for ai, bj in zip(a, b)]))
```

```python
a = np.array([10,10,10])
b = np.array([0,0,-10])
```

```python
vector_len(a, b)
```

```python
np.linalg.norm(a+b)
```

```python
def polar_decard(R, a):
    if R >= 0 and (0 <= a <= 2*np.pi):
        return R*np.cos(a), R*np.sin(a)
    else:
        print(' Введены неверные значения')
x,y = polar_decard(3, np.pi/3)
x, y
```

## Задание 4.1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5,5, 300)
plt.plot(x, x**2-1)
plt.plot(x, (np.exp(x)+x-1)/x)
```

```python
def equations(p):
    x, y = p
    return(y-x**2+1, np.exp(x)+x*(1-y)-1)
x1, y1 = fsolve(equations, (-2,2))
x1, y1
```

```python
x2, y2 = fsolve(equations, (3,6))
x2, y2
```

```python
x3, y3 = fsolve(equations, (4,15))
x3, y3
```

## Задание 4.2

```python
x_1 = np.linspace(x1, x2, 100)
x_2 = np.linspace(x3, 6, 100)
plt.plot(x_1, x_1**2-1, c='blue')
plt.plot(x_2, x_2**2-1, c='blue')
```
